```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import numpy as np
import random

height = 256
width = 256

image = np.zeros((height, width))

# fill first half with random noise at low intensity
for i in range(width//2):
    for j in range(height):
        image[j][i] = random.randint(0, 25)

# fill second half with random noise at lower intensity
for i in range(width//2, width):
    for j in range(height):
        image[j][i] = random.randint(25, 50)
        
for i in range(height//2, height):
    for j in range(width):
        image[i][j] = random.randint(25, 50)
        
        
plt.imshow(image, cmap='gray')
```

```python
# derivative of image
derivative = np.zeros((height, width))

for i in range(1, height-1):
    for j in range(1, width-1):
        derivative[i][j] = image[i][j] - image[i][j-1]
        
        
dx, dy = np.gradient(image)   

fig, axs = plt.subplots(1, 3, figsize=(10, 5))

axs[0].imshow(dy, cmap='gray')
axs[0].set_title('dx')

axs[1].imshow(dx, cmap='gray')
axs[1].set_title('dy')

axs[2].imshow(derivative, cmap='gray')
axs[2].set_title('my derivative')


plt.show()
```
